## 13.1 El Desafio la empresa

Cada equipo (2-3 personas) propone un caso de uso de IA  
que aporte valor real a la empresa o a sus clientes.

**Las 4 preguntas en este orden exacto:**

```
1. Que PROBLEMA resuelve
   -> proceso actual + ineficiencia + mejora medible

2. Que VARIABLE necesitas estimar
   -> magnitud concreta que el sistema debe producir
   -> sin esta respuesta, la selección de técnica carece de criterio

3. Que TECNOLOGIAS usa
   -> la justificacion PARTE de la variable, no de la preferencia

4. Como se MIDE el exito
   -> métricas concretas: tiempo, precisión, coste, satisfaccion
   -> sin métricas no hay forma de saber si la solucion funciona
```

**Catálogo por nivel de dificultad:**

| Nivel | Caso |

| Accesible | Clasificador de tickets (categoria + prioridad) |

````python
# Plantilla de propuesta del Desafio la empresa
# Rellenar y validar antes de empezar a construir

PROPUESTA_TEMPLATE = {
    'equipo': ['Nombre1', 'Nombre2'],
    'caso_de_uso': 'Titulo descriptivo del caso',

    '1_problema': {
        'situacion_actual': 'Como se hace hoy (proceso manual, tiempo invertido)',
        'dolor_principal': 'Que ineficiencia o limitacion tiene',
        'impacto_estimado': 'Que se gana al resolverlo (KPI concreto)',
    },

    '2_variable_objetivo': {
        'que_estima': 'Magnitud concreta que el sistema debe producir',
        'tipo_salida': 'numero continuo | categoria | estructura JSON | generacion',
        'por_que_no_con_reglas': 'Razon concreta de por que no es calculable con formulas',
    },

    '3_decision': {
        'usar_ia': 'si | no | parcial',
        'razon_una_frase': 'Por que si/no',
        'tipo_solucion': 'LLM via API | modelo tabular | RAG | agente | hibrido',
        'por_que_este_tipo': 'Criterio tecnico que descarta las alternativas',
    },

    '4_solucion': {
        'descripcion_funcional': 'Que hace la solucion en lenguaje de negocio',
        'arquitectura': 'Componentes tecnicos (LLM, RAG, pipeline, vector store...)',
        'modelo': 'Claude Haiku | Sonnet | XGBoost | otro',
    },

    '5_mvp': {
        'incluye': ['Funcionalidad minima 1', 'Funcionalidad minima 2'],
        'no_incluye': ['Limitacion consciente 1', 'Limitacion consciente 2'],
        'datos_necesarios': 'Que datos se necesitan y si estan disponibles',
    },

    '6_metricas': {
        'metrica_principal': 'precision > X% | latencia < Xs | ahorro Y EUR/mes',
        'metricas_secundarias': ['latencia por peticion', 'coste por peticion'],
        'criterio_aceptacion': 'Cuando se considera que funciona',
    },

    '7_plan': {
        'semana_1': 'Investigacion y diseno del prompt / arquitectura',
        'semana_2': 'Implementacion del core (notebook)',
        'semana_3': 'Dataset de test + evaluacion + ajustes',
        'semana_4': 'API + demo preparada para presentacion',
    },
}


def validar_propuesta(propuesta: dict) -> tuple:
    """Verifica que la propuesta tiene los campos criticos completos."""
    errores = []
    advertencias = []

    # Campo critico: variable objetivo
    var = propuesta.get('2_variable_objetivo', {})
    if 'TEMPLATE' in str(var.get('que_estima', '')) or not var.get('que_estima'):
        errores.append('La variable objetivo no esta definida concretamente')
    if var.get('tipo_salida', '') not in ['numero continuo', 'categoria', 'estructura JSON', 'generacion']:
        advertencias.append('El tipo de salida no es uno de los 4 tipos reconocidos')

    # Campo critico: metricas
    metricas = propuesta.get('6_metricas', {})
    if 'TEMPLATE' in str(metricas.get('metrica_principal', '')):
        errores.append('La metrica principal no tiene un valor numerico concreto')

    # Campo critico: datos disponibles
    mvp = propuesta.get('5_mvp', {})
    if 'disponibles' not in str(mvp.get('datos_necesarios', '')).lower():
        advertencias.append('Confirmar que los datos necesarios estan disponibles antes de empezar')

    return len(errores) == 0, errores, advertencias


# --- Ejemplo de propuesta bien completada ---
propuesta_ejemplo = dict(PROPUESTA_TEMPLATE)
propuesta_ejemplo['equipo'] = ['Ana Garcia', 'Pablo Martinez']
propuesta_ejemplo['caso_de_uso'] = 'Clasificador automatico de tickets de soporte'
propuesta_ejemplo['2_variable_objetivo'] = {
    'que_estima': 'Categoria (bug/feature/pregunta/config) y prioridad (critica/alta/media/baja) de cada ticket',
    'tipo_salida': 'categoria',
    'por_que_no_con_reglas': 'El lenguaje natural de los tickets es ambiguo y variable; reglas de palabras clave fallan en el 30% de los casos',
}
propuesta_ejemplo['6_metricas'] = {
    'metrica_principal': 'precision de categoria > 90% en dataset de test de 100 tickets',
    'metricas_secundarias': ['latencia < 2s por ticket', 'coste < 0.001 USD por ticket'],
    'criterio_aceptacion': 'precision >= 90% y latencia < 2s en el 95% de los casos',
}

ok, errores, advertencias = validar_propuesta(propuesta_ejemplo)
print('=== Validacion de la propuesta del equipo ===')
print(f'Equipo: {propuesta_ejemplo["equipo"]}')
print(f'Caso:   {propuesta_ejemplo["caso_de_uso"]}')
print()
if ok:
    print('[OK] La propuesta supera la validacion minima')
else:
    print('[X]  La propuesta tiene campos criticos sin completar')
    for e in errores:
        print(f'  ERROR: {e}')
for a in advertencias:
    print(f'  [!] Advertencia: {a}')
````

## 13.2 Estructura del MVP

**Stack recomendado para el prototipo:**

```
Backend:      Python + FastAPI
LLM:          API de Claude (Anthropic) - claude-haiku para clasificacion,
              claude-sonnet para generación compleja
Vector Store: ChromaDB (prototipo rápido de RAG)
ML clásico:   Scikit-learn o XGBoost para modelos tabulares
Frontend:     Streamlit o Gradio (demo visual opcional)
```

**Ciclo de desarrollo:**
```
1. Notebook: implementar el core, iterar prompts con ejemplos manuales
2. Dataset de test: 50-100 ejemplos etiquetados por un experto del dominio
3. Evaluación: ejecutar el notebook contra el dataset, medir métricas
4. API (FastAPI): empaquetar con validación de entrada, manejo de errores, logging
5. Integración (opcional): conectar con MCP o endpoint de la empresa
```

````python
# Estructura de un MVP con FastAPI
# Este codigo muestra la arquitectura - no ejecuta una API real en el notebook

CODIGO_FASTAPI_MVP = '''
from fastapi import FastAPI, HTTPException
from pydantic import BaseModel
import anthropic, json, logging

app = FastAPI(title='Sistema AI - Clasificador de Tickets')
client = anthropic.Anthropic()  # usa ANTHROPIC_API_KEY del entorno

class TicketInput(BaseModel):
    titulo: str
    descripcion: str
    modulo: str

class TicketOutput(BaseModel):
    categoria: str
    prioridad: str
    confianza: float
    sugerencia: str

PROMPT_TEMPLATE = """
Eres un analista de soporte de nivel 2 de la empresa.
Clasifica el siguiente ticket.

Modulo: {modulo}
Titulo: {titulo}
Descripcion: {descripcion}

Responde en JSON:
{{"categoria": "bug|feature|pregunta|config",
  "prioridad": "critica|alta|media|baja",
  "confianza": 0.0-1.0,
  "sugerencia": "accion recomendada en una frase"}}
"""

@app.post("/clasificar", response_model=TicketOutput)
async def clasificar_ticket(ticket: TicketInput):
    prompt = PROMPT_TEMPLATE.format(
        modulo=ticket.modulo,
        titulo=ticket.titulo,
        descripcion=ticket.descripcion,
    )
    response = client.messages.create(
        model="claude-haiku-4-5-20251001",
        max_tokens=300,
        messages=[{"role": "user", "content": prompt}],
    )
    try:
        resultado = json.loads(response.content[0].text)
        return TicketOutput(**resultado)
    except (json.JSONDecodeError, KeyError) as e:
        raise HTTPException(status_code=422, detail=f"Respuesta invalida del modelo: {e}")

@app.get("/health")
def health():
    return {"status": "ok", "model": "claude-haiku-4-5-20251001"}
'''

print('=== Estructura del MVP FastAPI ===')
print()
print('Endpoints expuestos:')
print('  POST /clasificar  -> clasifica un ticket y retorna JSON estructurado')
print('  GET  /health      -> verifica que el servicio esta disponible')
print()
print('Para ejecutar localmente:')
print('  pip install fastapi uvicorn anthropic')
print('  export ANTHROPIC_API_KEY=sk-ant-...')
print('  uvicorn main:app --reload')
print()
print('Codigo del MVP (ver variable CODIGO_FASTAPI_MVP):')
print(f'  {len(CODIGO_FASTAPI_MVP.splitlines())} lineas | {len(CODIGO_FASTAPI_MVP)} chars')
print()
print('[OK] Un MVP funcional cabe en ~50 lineas de Python.')
print('La complejidad viene de la evaluacion y los guardrails, no del codigo base.')
````

## 13.2 Evaluación del MVP

**Evaluación técnica** (métricas clásicas para clasificación):
- Precisión, recall y F1-score por categoria
- Latencia (P50, P95) y coste por peticion
- Robustez ante entradas inesperadas o adversariales

**Evaluación de negocio:**
- Resuelve el problema identificado?
- El ahorro de tiempo/coste justifica el coste de la solución?
- Es adoptable sin formación adicional significativa?
- Es mantenible a largo plazo?

**LLM-as-Judge** (para tareas de generación donde no hay respuesta única correcta):  
Un modelo mas potente evalua la calidad de las respuestas del sistema bajo prueba.  
Umbral recomendado para la empresa: 7/10 con muestreo del 10% de las respuestas de producción.

````python
# Framework de evaluacion automatizada del MVP

def evaluar_clasificador(clasificador_fn, dataset_test: list) -> dict:
    """
    Evalua un clasificador contra un dataset de test.
    Calcula precision global y por categoria.
    """
    total = len(dataset_test)
    correctos = 0
    por_categoria = {}
    errores = []

    for ejemplo in dataset_test:
        prediccion = clasificador_fn(ejemplo['titulo'], ejemplo['descripcion'])
        cat_real = ejemplo['categoria_real']
        cat_pred = prediccion.get('categoria', 'desconocido')

        if cat_real not in por_categoria:
            por_categoria[cat_real] = {'total': 0, 'correctos': 0}
        por_categoria[cat_real]['total'] += 1

        if cat_pred == cat_real:
            correctos += 1
            por_categoria[cat_real]['correctos'] += 1
        else:
            errores.append({'titulo': ejemplo['titulo'], 'esperado': cat_real, 'obtenido': cat_pred})

    precision_global = correctos / total if total > 0 else 0
    for cat, datos in por_categoria.items():
        datos['precision'] = datos['correctos'] / datos['total'] if datos['total'] > 0 else 0

    return {
        'precision_global': precision_global,
        'correctos': correctos,
        'total': total,
        'por_categoria': por_categoria,
        'errores': errores,
    }


# Clasificador simulado (en produccion: llamada al LLM)
def clasificador_simulado(titulo: str, descripcion: str) -> dict:
    """Modelo simulado para demostracion de la evaluacion."""
    texto = (titulo + ' ' + descripcion).lower()
    if any(w in texto for w in ['error', 'fallo', 'no funciona', 'sale', 'bug']):
        return {'categoria': 'bug', 'prioridad': 'alta', 'confianza': 0.88}
    if any(w in texto for w in ['gustaria', 'podria', 'nueva funcionalidad', 'añadir']):
        return {'categoria': 'feature', 'prioridad': 'media', 'confianza': 0.82}
    if any(w in texto for w in ['como', 'donde', 'es posible', 'se puede']):
        return {'categoria': 'pregunta', 'prioridad': 'baja', 'confianza': 0.79}
    if any(w in texto for w in ['configurar', 'permiso', 'acceso', 'rol']):
        return {'categoria': 'config', 'prioridad': 'alta', 'confianza': 0.84}
    return {'categoria': 'pregunta', 'prioridad': 'baja', 'confianza': 0.5}


# Dataset de test (en produccion: 50-100 tickets etiquetados por un experto)
DATASET_TEST = [
    {'titulo': 'Error al exportar informe',     'descripcion': 'Sale un error 500 al pulsar exportar', 'categoria_real': 'bug'},
    {'titulo': 'No funciona el filtro de fechas','descripcion': 'El filtro de fechas no responde',      'categoria_real': 'bug'},
    {'titulo': 'Descuentos por volumen',         'descripcion': 'Me gustaria poder configurar descuentos automaticos', 'categoria_real': 'feature'},
    {'titulo': 'Integracion con ERP',            'descripcion': 'Podria conectarse con nuestro SAP B1?', 'categoria_real': 'feature'},
    {'titulo': 'Como cambiar el proveedor',      'descripcion': 'Es posible cambiar el proveedor de un pedido ya confirmado?', 'categoria_real': 'pregunta'},
    {'titulo': 'Donde se configura el IVA',     'descripcion': 'No encuentro donde configurar los tipos de IVA', 'categoria_real': 'pregunta'},
    {'titulo': 'Permisos del rol Admin',         'descripcion': 'El rol Admin no tiene acceso al modulo de facturacion', 'categoria_real': 'config'},
    {'titulo': 'Acceso denegado al dashboard',   'descripcion': 'Un usuario con rol Viewer no puede ver el dashboard', 'categoria_real': 'config'},
]

resultado = evaluar_clasificador(clasificador_simulado, DATASET_TEST)

print(f'=== Evaluacion del Clasificador de Tickets ===')
print(f'Dataset: {resultado["total"]} ejemplos | Correctos: {resultado["correctos"]}')
print(f'Precision global: {resultado["precision_global"]*100:.1f}%')
print()
print('Precision por categoria:')
for cat, datos in resultado['por_categoria'].items():
    barra = '#' * int(datos['precision'] * 20)
    print(f'  {cat:<10} {datos["precision"]*100:5.1f}%  [{barra:<20}]  ({datos["correctos"]}/{datos["total"]})')

umbral = 0.85
supera = resultado['precision_global'] >= umbral
print()
print(f'Umbral de aceptacion: {umbral*100:.0f}%')
estado = '[OK] SUPERA el umbral' if supera else '[X]  NO supera el umbral'
print(f'{estado}')

if resultado['errores']:
    print(f'\nErrores para revisar ({len(resultado["errores"])}):')
    for e in resultado['errores']:
        print(f'  Titulo: "{e["titulo"]}" | Esperado: {e["esperado"]} | Obtenido: {e["obtenido"]}')
````

````python
# LLM-as-Judge: evaluacion de respuestas generativas
# Para tareas como User Stories, borradores de propuesta, respuestas de RAG

CRITERIOS_USER_STORY = [
    {'nombre': 'formato_correcto',    'descripcion': 'Usa formato "Como [rol], quiero [accion], para [beneficio]"', 'max': 2},
    {'nombre': 'criterios_giwt',      'descripcion': 'Criterios en formato Given/When/Then (minimo 3)', 'max': 3},
    {'nombre': 'sin_ambiguedad',       'descripcion': 'Sin terminos ambiguos como "facil", "rapido", "bueno"', 'max': 2},
    {'nombre': 'contexto_empresa',  'descripcion': 'Especifico para el contexto de la empresa', 'max': 2},
    {'nombre': 'estimable',            'descripcion': 'El equipo de dev puede estimar la complejidad', 'max': 1},
]

def evaluar_user_story_simulado(user_story: str) -> dict:
    """
    Evalua una User Story con criterios estructurados.
    En produccion: llamada al LLM evaluador (claude-sonnet o similar).
    """
    puntuaciones = {}
    total = 0

    us_lower = user_story.lower()

    # Formato
    tiene_formato = 'como ' in us_lower and 'quiero ' in us_lower and 'para ' in us_lower
    puntuaciones['formato_correcto'] = 2 if tiene_formato else 0

    # Given/When/Then
    n_gwt = us_lower.count('dado') + us_lower.count('cuando') + us_lower.count('entonces') + us_lower.count('given') + us_lower.count('when') + us_lower.count('then')
    puntuaciones['criterios_giwt'] = min(3, n_gwt // 2)

    # Sin ambiguedad
    ambiguos = ['facil', 'rapido', 'bueno', 'mejor', 'correcto', 'adecuado', 'optimo']
    n_ambiguos = sum(1 for a in ambiguos if a in us_lower)
    puntuaciones['sin_ambiguedad'] = 0 if n_ambiguos > 1 else 1 if n_ambiguos == 1 else 2

    # Contexto la empresa
    terminos_contexto = ['la empresa', 'inventario', 'pedido', 'almacen', 'cliente', 'producto', 'factura']
    n_contexto = sum(1 for t in terminos_contexto if t in us_lower)
    puntuaciones['contexto_empresa'] = min(2, n_contexto)

    # Estimable
    puntuaciones['estimable'] = 1 if len(user_story) > 200 else 0

    total = sum(puntuaciones.values())
    max_total = sum(c['max'] for c in CRITERIOS_USER_STORY)

    return {'puntuaciones': puntuaciones, 'total': total, 'max': max_total, 'ratio': total / max_total}


# Dos User Stories para comparar
US_MALA = """
Como usuario, quiero poder gestionar el inventario de forma mas facil y rapida,
para que sea mejor.
"""

US_BUENA = """
Como responsable de almacen de la empresa, quiero recibir una alerta automatica
cuando el stock de un producto cae por debajo del punto de pedido configurado,
para poder lanzar la orden de compra antes de que se produzca una rotura de stock.

Criterios de aceptacion:
- Dado que el stock del producto REF-4521 esta en 25 unidades y el punto de pedido es 30,
  Cuando el sistema procesa el movimiento de salida,
  Entonces se envia una notificacion al responsable de almacen via email y en el panel de alertas.
- Dado que el producto tiene el calculo automatico desactivado,
  Cuando el stock cae por debajo del umbral,
  Entonces NO se envia ninguna alerta (respeta la configuracion).
- Dado que el responsable recibe la alerta,
  Cuando hace clic en el enlace de la notificacion,
  Entonces accede directamente al formulario de orden de compra con el producto preseleccionado.
"""

print('=== LLM-as-Judge: Evaluacion de User Stories ===')
for nombre, us in [('User Story basica', US_MALA), ('User Story completa', US_BUENA)]:
    resultado = evaluar_user_story_simulado(us)
    ratio = resultado['ratio']
    nivel = 'EXCELENTE' if ratio >= 0.85 else 'BUENA' if ratio >= 0.65 else 'MEJORABLE' if ratio >= 0.40 else 'INSUFICIENTE'
    print(f'\n--- {nombre} ---')
    print(f'Puntuacion: {resultado["total"]}/{resultado["max"]} ({ratio*100:.0f}%) [{nivel}]')
    for criterio in CRITERIOS_USER_STORY:
        k = criterio['nombre']
        v = resultado['puntuaciones'][k]
        m = criterio['max']
        estado = '[OK]' if v == m else '[!] ' if v > 0 else '[X] '
        print(f'  {estado} {k:<22} {v}/{m}')
````

## 13.4 Estrategia IA First de la empresa

**IA First no significa que todo deba resolverse con IA.**  
Significa que para cada decisión de producto, arquitectura o proceso,  
la primera pregunta es: *¿Cómo puede la IA mejorar esto?*  
La respuesta puede ser 'no puede' o 'no vale la pena', y eso esta bien.

**Roadmap de integración:**

```
FASE 1 - Quick Wins (0-3 meses)
 - Clasificación automática de tickets de soporte
 - Búsqueda semántica sobre documentación interna
 - Asistente de prompting para el equipo (plantillas + best practices)

FASE 2 - Integración (3-6 meses)
 - RAG sobre base de conocimiento de la empresa para soporte nivel 1
 - Copiloto para Product Owner integrado en flujo de trabajo
 - Pipeline de análisis de sentimiento para Customer Success

FASE 3 - Diferenciación (6-12 meses)
 - Predicción de inventarios como funcionalidad de la empresa para clientes
 - Procesamiento inteligente de RFPs en el flujo comercial
 - Sistema multi-agente para automatización de procesos internos

FASE 4 - Plataforma (12+ meses)
 - Infraestructura MCP que conecta todos los servicios de la empresa
 - Capacidades de IA expuestas como APIs que los clientes personalizan
 - Ecosistema de agentes colaborativos que aprenden con el uso
```

````python
# Calculadora de ROI para el roadmap de integracion

import matplotlib.pyplot as plt

INICIATIVAS = [
    {
        'nombre': 'Clasificador tickets',
        'fase': 1,
        'coste_desarrollo_h': 80,
        'coste_api_mes_eur': 10,
        'ahorro_mensual_eur': 2000,
        'coste_hora_dev': 60,
    },
    {
        'nombre': 'RAG documentacion',
        'fase': 2,
        'coste_desarrollo_h': 160,
        'coste_api_mes_eur': 40,
        'ahorro_mensual_eur': 3500,
        'coste_hora_dev': 60,
    },
    {
        'nombre': 'Copiloto PO',
        'fase': 2,
        'coste_desarrollo_h': 120,
        'coste_api_mes_eur': 25,
        'ahorro_mensual_eur': 2800,
        'coste_hora_dev': 60,
    },
    {
        'nombre': 'Prediccion inventarios',
        'fase': 3,
        'coste_desarrollo_h': 400,
        'coste_api_mes_eur': 150,
        'ahorro_mensual_eur': 8000,
        'coste_hora_dev': 70,
    },
    {
        'nombre': 'Procesador RFPs',
        'fase': 3,
        'coste_desarrollo_h': 300,
        'coste_api_mes_eur': 80,
        'ahorro_mensual_eur': 5000,
        'coste_hora_dev': 70,
    },
]

print('=== ROI por iniciativa (proyeccion a 12 meses) ===')
print()
print(f'{"Iniciativa":<25} {"Inversion":<12} {"Ahorro/ano":<12} {"ROI":<8} {"Mes BE"}')
print('-' * 70)

for ini in INICIATIVAS:
    inversion = ini['coste_desarrollo_h'] * ini['coste_hora_dev'] + ini['coste_api_mes_eur'] * 12
    ahorro_anual = ini['ahorro_mensual_eur'] * 12
    roi = (ahorro_anual - inversion) / inversion if inversion > 0 else 0
    # Mes de break-even
    coste_dev = ini['coste_desarrollo_h'] * ini['coste_hora_dev']
    margen_mensual = ini['ahorro_mensual_eur'] - ini['coste_api_mes_eur']
    be_mes = int(coste_dev / margen_mensual) + 1 if margen_mensual > 0 else 999

    estado = '[OK]' if roi > 0 else '[!] '
    print(f'{estado} {ini["nombre"]:<22} {inversion:>8,.0f} EUR  {ahorro_anual:>8,.0f} EUR  {roi*100:>5.0f}%  mes {be_mes}')

print()
total_inversion = sum(i['coste_desarrollo_h']*i['coste_hora_dev'] + i['coste_api_mes_eur']*12 for i in INICIATIVAS)
total_ahorro = sum(i['ahorro_mensual_eur']*12 for i in INICIATIVAS)
print(f'TOTAL ROADMAP: inversion {total_inversion:,.0f} EUR | ahorro anual {total_ahorro:,.0f} EUR')
print(f'ROI global del roadmap: {(total_ahorro - total_inversion)/total_inversion*100:.0f}%')
````
